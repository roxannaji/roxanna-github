# Create global functions

```python
# Create a function for further use: output a list of matched PVs based on the input currencies list
def fd_pv(df, currencies):
    pv_lst = []
    for currency in currencies:
        pv = df.loc[df['currency'] == currency, 'PV'].values
        pv_lst.append(pv)
    return pv_lst
```

```python
CurrCode = ['USD', 'EUR', 'GBP', 'CHF', 'JPY', 'AUD', 'NZD', 'ILS', 'NOK', 
                'SEK', 'SGD', 'DKK', 'HKD', 'CAD']
fd_pv(df_report, CurrCode)
```

```python
# Create a function for further use: convert a column of data whose types are list to types of float
def update_column(df, column_name):
    df[column_name] = df[column_name].apply(lambda x: x[0])
    return df
```

```python
# Create a function for further use: Find hedge ratio and update the df with the calculated hedgeable exposure
def fd_ratio(df, strategy):
    index = df.index
    values = [0.0 for _ in range(len(index))]
    df['ratio'] = values
    for i in index:
        if (i in df_hedge_mapping['Currency'].values):
            df_filtered = df_hedge_mapping[(df_hedge_mapping['IPS Strategy'] == strategy)]
            df.at[i, 'ratio'] = df_filtered.loc[df_filtered['Currency'] == i,'Hedge Ratio'].sum()
    df['Hedgeable Exposure'] = df['ratio'] * df['Underlying Exposure']
    df = df.drop(columns=['ratio'])
    return(df)
```

```python
# Calculate scaling ratio needed for calculation in BM exposure
def scale_ratio(client, strategy, curr):
    df_fil = df_bm[
        (df_bm["client"] == client) &
        (df_bm["IPS_Strategy"] == strategy)]
    df_fil1 = df_fil[df_fil['currency_new'] == curr]
    
    a = df_fil1['exposure'].sum()
    b = df_fil['exposure'].sum()
    
    if b != 0:  
        weight = a/b
        return weight
    else: 
        return 0


# Testing
print(scale_ratio('OPB',"Global Infrastructure",'USD'))
print(scale_ratio('OPB',"Global Real Estate",'GBP'))
```

```python
# Function: Sum up all the values of given currencies for a specified column
def sum_up(df, col):
    lst1 = ['USD', 'EUR', 'GBP', 'CHF', 'JPY', 'AUD', 'NZD', 'ILS', 'NOK','SEK', 'SGD', 'DKK', 'HKD', 'CAD']
    lst2 = ['USD', 'EUR', 'GBP', 'CHF', 'JPY', 'AUD', 'NZD', 'ILS', 'NOK','SEK', 'SGD', 'DKK', 'HKD']
    val = 0
    if col == 'Strategic Hedges':
        for i in lst2:  
            num = df[col][i]
            val = val + num
    else:
        for i in lst1:
            num = df[col][i]
            val = val + num
            
    return val
```

```python
# Function: Create tables for a given Strategy and Client
def FX_Exp(client, strategy):
    # Filter the data based on the specified criteria
    df_filtered = df_summary[
        (df_summary["client"] == client) &
        (df_summary["IPS_Strategy"] == strategy)]
    df_filtered1 = df_filtered[(df_filtered["Hedge Account"] == False)]
    df_filtered2 = df_filtered[(df_filtered["Hedge Account"] == True)]
    df_pool = df_filtered1[(df_filtered1["PortfolioParameter"] == 'GVCC0005')]
    df_seg = df_filtered1[df_filtered1["PortfolioParameter"] != 'GVCC0005']
 
    # Set up dataframe
    columns = ['Underlying Exposure (Local)','Underlying Exposure', 'Hedgeable Exposure', 'Benchmark Exposure','Strategic Hedges']
    CurrCode = ['USD', 'EUR', 'GBP', 'CHF', 'JPY', 'AUD', 'NZD', 'ILS', 'NOK', 
                'SEK', 'SGD', 'DKK', 'HKD', 'CAD']
    output = pd.DataFrame(columns=columns, index = CurrCode)

    # Update Underlying Exposure
      
    for i in CurrCode:
        output.at[i, 'Underlying Exposure'] = df_filtered1.loc[df_filtered1['currency_new'] == i,'exposure'].sum()/1000000  

    # Update Active Currency if strategy == Public Market Alternatives
    if(strategy == 'Public Market Alternatives'):
        output.insert(2, 'Active Currency', 0.0)
        for i in CurrCode:
            output.at[i, 'Active Currency'] = df_filtered1.query('currency_new == @i and IMCO_PortfolioID == "IMP052C"')['exposure'].sum()/1000000 
        output['Underlying Exposure'] = output['Underlying Exposure'] - output['Active Currency']
    
    # Update Hedgeable Exposure
    output = fd_ratio(output, strategy)

    # Update Strategic Hedges
    for i in CurrCode:
        output.at[i, 'Strategic Hedges'] = df_filtered2.loc[df_filtered2['currency_new'] == i,'exposure'].sum()/1000000    
    output.at['CAD', 'Strategic Hedges'] = 0
    sum_row = output.sum()
    
    # Update Underlying Exposure (Local)
    PVs = fd_pv(df_report, CurrCode) # PVs is a list of PVs corresponding to each value in CurrCode
    output['Underlying Exposure (Local)'] = output['Underlying Exposure'].div(PVs)
    update_column(output, 'Underlying Exposure (Local)')
    
    # Total DM for columns except 'Benchmark'
    output.loc['Total DM'] = sum_row
    
    # Total EM for columns except 'Benchmark'
    output.loc['Total EM'] = None
    output.at['Total EM', 'Underlying Exposure (Local)'] = pd.NA
    total = df_filtered1['exposure'].sum()/1000000 
    total_dm = output.at['Total DM','Underlying Exposure']
    total_em = total - total_dm
    output.at['Total EM', 'Underlying Exposure'] = total_em
    output.at['Total EM', 'Hedgeable Exposure'] = total_em

    
    # Update Benchmark Exposure
    for i in CurrCode:
        bm = total * scale_ratio(client, strategy, i)
        output.at[i, 'Benchmark Exposure'] = bm
    
    # Total DM and EM for Benchmark Exposure
    output.at['Total DM','Benchmark Exposure'] = output['Benchmark Exposure'].sum()
    bm_dm = output.at['Total DM','Benchmark Exposure']
    bm_em = total - bm_dm
    output.at['Total EM','Benchmark Exposure'] = bm_em

    # Use float as data types and round to 1 decimal
    output = output.astype(float)
    #output = output.round(1)
    
    # Update NA
    output.at['CAD', 'Strategic Hedges'] = pd.NA
    output.at['Total DM', 'Underlying Exposure (Local)'] = pd.NA
    output.at['Total EM', 'Underlying Exposure (Local)'] = pd.NA
    output.at['Total EM', 'Strategic Hedges'] = pd.NA
    
    df_fil = df_summary[
        (df_summary["IPS_Strategy"] == strategy) &
        (df_summary["Hedge Account"] == False)]

    
    df_filtered1 = df_filtered[(df_filtered["Hedge Account"] == False)]
   
    # For the second tab FX Exposure - Pool -- Update column of Active Currency Exposure
     
    if (client == "IMCO") and (strategy == 'Public Market Alternatives'):
        output.insert(2, 'Active Currency Exposure', 0.0)
        for i in CurrCode:
            output.at[i, 'Active Currency Exposure'] = df_fil.query('currency_new == @i and IMCO_PortfolioID == "IMP052C"')['exposure'].sum()/1000000     
        output.at['Total DM','Active Currency Exposure'] = output['Active Currency Exposure'].sum()

        df_filtered3 = df_filtered1[(df_filtered["IMCO_PortfolioID"] == "IMP052C")]
        output.at['Total EM', 'Active Currency Exposure'] = df_filtered1.query('IMCO_PortfolioID == "IMP052C"')['exposure'].sum()/1000000 - output.at['Total DM','Active Currency Exposure']
        output = output.drop('Active Currency', axis=1)
    
    # Caclulation for exception case: client = IMCO
    if (client == "IMCO"):
        tot = output.at['Total EM', 'Underlying Exposure'] + output.at['Total DM', 'Underlying Exposure']
        for i in CurrCode:
            bm = tot * scale_ratio("OPB", strategy, i)
            output.at[i, 'Benchmark Exposure'] = bm
          
    
        bmdm = sum_up(output, 'Benchmark Exposure')
        bmem = tot - bmdm
        output.at['Total DM','Benchmark Exposure'] = bmdm
        output.at['Total EM','Benchmark Exposure'] = bmem
  
    
    
    return output.shape[1], output
```

```python
# Print out the table for the first tab: FX Exposure - Client 
def print_table(client, start_row, start_col):
    # Print out date
    with pd.ExcelWriter(new_file_name, engine = 'openpyxl', mode = 'a', if_sheet_exists = 'overlay') as writer:
        # Step 2: Create a DataFrame from the variable
        time_df = pd.DataFrame({'Value': [date]})
        time_df.to_excel(writer, sheet_name = 'FX Exposure - Client', startrow = start_row - 4, startcol = start_col + 1, index = False, header = False)
            
    # Print dataframe
    strategies = ["Public Market Alternatives", "Inflation Linked Bond",
              "Total Portfolio Leverage", "Global Credit", "Money Market"]
    with pd.ExcelWriter(new_file_name, engine = 'openpyxl', mode = 'a', if_sheet_exists = 'overlay') as writer:
        for i in range(len(strategies)):
            col, output = FX_Exp(client, strategies[i])
            # Empty the template table
            new_index = range(len(output) + 1)
            empty_df = pd.DataFrame(index = new_index, columns = output.columns)
            empty_df.to_excel(writer, sheet_name = 'FX Exposure - Client', startrow = start_row, startcol =start_col, index = False, header = False)
            output.to_excel(writer, sheet_name = 'FX Exposure - Client', startrow = start_row, startcol = start_col, index = False, header = False)
            start_col += col
```

```python
def FX_first3(client, strategy, re): #re=0 if strategy is GI; re=pool if strategy is GRE-Pool
    df_filtered = df_summary[
        (df_summary["client"] == client) &
        (df_summary["IPS_Strategy"] == strategy)]
    df_filtered1 = df_filtered[(df_filtered["Hedge Account"] == False)]
    df_filtered2 = df_filtered[(df_filtered["Hedge Account"] == True)]
    df_pool = df_filtered1[(df_filtered1["PortfolioParameter"] == 'GVCC0005')]
    df_seg = df_filtered1[df_filtered1["PortfolioParameter"] != 'GVCC0005']
 
    # Set up dataframe
    columns = ['Underlying Exposure (Local)','Underlying Exposure', 'Hedgeable Exposure','Benchmark Exposure','Strategic Hedges']
    CurrCode = ['USD', 'EUR', 'GBP', 'CHF', 'JPY', 'AUD', 'NZD', 'ILS', 'NOK', 
                'SEK', 'SGD', 'DKK', 'HKD', 'CAD']
    output = pd.DataFrame(columns=columns, index = CurrCode)

    if (re == 'pool'):
        for i in CurrCode:
            output.at[i, 'Underlying Exposure'] = df_pool.loc[df_pool['currency_new'] == i,'exposure'].sum()/1000000  
    if (re == 'seg'):
        for i in CurrCode:
            a = df_seg.loc[df_seg['currency_new'] == i,'exposure'].sum()/1000000 
            b = df_pool.loc[df_pool['currency_new'] == i,'exposure'].sum()/1000000 
            output.at[i, 'Underlying Exposure'] = a - b
    if (re == 0) and (strategy == 'Global Infrastructure'):
        for i in CurrCode:
            output.at[i, 'Underlying Exposure'] = df_filtered1.loc[df_filtered1['currency_new'] == i,'exposure'].sum()/1000000  
    if (re == 0) and (strategy == 'Global Real Estate'):
        for i in CurrCode:
            output.at[i, 'Underlying Exposure'] = df_seg.loc[df_seg['currency_new'] == i,'exposure'].sum()/1000000  
    
    # Update Hedgeable Exposure
    output = fd_ratio(output, strategy)

    # Update Strategic Hedges
    for i in CurrCode:
        output.at[i, 'Strategic Hedges'] = df_filtered2.loc[df_filtered2['currency_new'] == i,'exposure'].sum()/1000000    
    output.at['CAD', 'Strategic Hedges'] = 0
    sum_row = output.sum()
    
    # Update Underlying Exposure (Local)
    PVs = fd_pv(df_report, CurrCode) # PVs is a list of PVs corresponding to each value in CurrCode
    output['Underlying Exposure (Local)'] = output['Underlying Exposure'].div(PVs)
    update_column(output, 'Underlying Exposure (Local)')
    
    # Total DM for columns except 'Benchmark'
    output.loc['Total DM'] = sum_row
    
    # Total EM for columns except 'Benchmark'
    output.loc['Total EM'] = None
    output.at['Total EM', 'Underlying Exposure (Local)'] = pd.NA
    if (re == 'pool'): 
        total = df_pool['exposure'].sum()/1000000 
        total_dm = output.at['Total DM','Underlying Exposure']
        total_em = total - total_dm
        output.at['Total EM', 'Underlying Exposure'] = total_em
        output.at['Total EM', 'Hedgeable Exposure'] = total_em
    
        # Update Benchmark Exposure
        for i in CurrCode:
            bm = total * scale_ratio(client, strategy, i)
            output.at[i, 'Benchmark Exposure'] = bm
            
        output.at['Total DM','Benchmark Exposure'] = output['Benchmark Exposure'].sum()
        bm_dm = output.at['Total DM','Benchmark Exposure']
        bm_em = total - bm_dm
        output.at['Total EM','Benchmark Exposure'] = bm_em
        
############################ Look at seg in a separate way ################################################  

    if (re == 'seg'):
        pool = df_pool['exposure'].sum()/1000000 
        nonpool = df_seg['exposure'].sum()/1000000 
        segregate = nonpool - pool
        
        output.at['Total DM','Underlying Exposure'] = sum_up(output,'Underlying Exposure')
        total_dm = output.at['Total DM','Underlying Exposure']
        
        total_em = nonpool - pool - total_dm
        output.at['Total EM', 'Underlying Exposure'] = total_em
        output.at['Total EM', 'Hedgeable Exposure'] = total_em
    
        # Update Benchmark Exposure
        for i in CurrCode:
            
            bm = segregate * scale_ratio(client, strategy, i)
            output.at[i, 'Benchmark Exposure'] = bm
            
        output.at['Total DM','Benchmark Exposure'] = sum_up(output,'Benchmark Exposure')
        bm_dm = output.at['Total DM','Benchmark Exposure']
        bm_em = segregate - bm_dm
        output.at['Total EM','Benchmark Exposure'] = bm_em  


    # Use float as data types and round to 1 decimal
    output = output.astype(float)
    #output = output.round(1)
    
    # Update NA
    output.at['CAD', 'Strategic Hedges'] = pd.NA
    output.at['Total DM', 'Underlying Exposure (Local)'] = pd.NA
    output.at['Total EM', 'Underlying Exposure (Local)'] = pd.NA
    output.at['Total EM', 'Strategic Hedges'] = pd.NA
  

    if (re == 0):
        nonpool = df_seg['exposure'].sum()/1000000 
        output.at['Total DM','Underlying Exposure'] = sum_up(output,'Underlying Exposure')
        total_dm = output.at['Total DM','Underlying Exposure']
        
        total_em = nonpool - total_dm
        output.at['Total EM', 'Underlying Exposure'] = total_em
        output.at['Total EM', 'Hedgeable Exposure'] = total_em
    

        for i in CurrCode:
            bm = nonpool * scale_ratio(client, strategy, i)
            output.at[i, 'Benchmark Exposure'] = bm
            
        output.at['Total DM','Benchmark Exposure'] = sum_up(output,'Benchmark Exposure')
        bm_dm = output.at['Total DM','Benchmark Exposure']
        bm_em = nonpool - bm_dm
        output.at['Total EM','Benchmark Exposure'] = bm_em  

    # Caclulation for exception case: client = IMCO
    if (client == "IMCO"):
        tot = output.at['Total EM', 'Underlying Exposure'] + output.at['Total DM', 'Underlying Exposure']
        for i in CurrCode:
            bm = tot * scale_ratio("OPB", strategy, i)
            output.at[i, 'Benchmark Exposure'] = bm
          
    
        bmdm = sum_up(output, 'Benchmark Exposure')
        bmem = tot - bmdm
        output.at['Total DM','Benchmark Exposure'] = bmdm
        output.at['Total EM','Benchmark Exposure'] = bmem
    
    return output
```

```python
# Function: Create similar tables for a given Strategy and Client
def FX_Exp3(client, strategy):
    # Filter the data based on the specified criteria
    df_filtered = df_summary[
        (df_summary["client"] == client) &
        (df_summary["IPS_Strategy"] == strategy)]
    df_filtered1 = df_filtered[(df_filtered["Hedge Account"] == False)]
    df_filtered2 = df_filtered[(df_filtered["Hedge Account"] == True)]

    # Set up dataframe
    columns = ['Underlying Exposure', 'Benchmark Exposure','Strategic Hedges']
    CurrCode = ['USD', 'EUR', 'GBP', 'CHF', 'JPY', 'AUD', 'NZD', 'ILS', 'NOK', 
                'SEK', 'SGD', 'DKK', 'HKD', 'CAD']
    output = pd.DataFrame(columns=columns, index = CurrCode)

    # Update Underlying Exposure
    for i in CurrCode:
        output.at[i, 'Underlying Exposure'] = df_filtered1.loc[df_filtered1['currency_new'] == i,'exposure'].sum()/1000000  

    # Update Active Currency if strategy == Public Market Alternatives
    if(strategy == 'Public Market Alternatives'):
        output.insert(2, 'Active Currency', 0.0)
        for i in CurrCode:
            output.at[i, 'Active Currency'] = df_filtered1.query('currency_new == @i and IMCO_PortfolioID == "IMP052C"')['exposure'].sum()/1000000 
        output['Underlying Exposure'] = output['Underlying Exposure'] - output['Active Currency']
    
    # Update Hedgeable Exposure
    output = fd_ratio(output, strategy)

    # Update Strategic Hedges
    for i in CurrCode:
        output.at[i, 'Strategic Hedges'] = df_filtered2.loc[df_filtered2['currency_new'] == i,'exposure'].sum()/1000000    
    output.at['CAD', 'Strategic Hedges'] = 0
    sum_row = output.sum()
    
    
    # Total DM for columns except 'Benchmark'
    output.loc['Total DM'] = sum_row
    
    # Total EM for columns except 'Benchmark'
    output.loc['Total EM'] = None
    
    total = df_filtered1['exposure'].sum()/1000000 
    total_dm = output.at['Total DM','Underlying Exposure']
    total_em = total - total_dm
    output.at['Total EM', 'Underlying Exposure'] = total_em

    
    # Update Benchmark Exposure
    for i in CurrCode:
        bm = total * scale_ratio(client, strategy, i)
        output.at[i, 'Benchmark Exposure'] = bm
    
    # Total DM and EM for Benchmark Exposure
    output.at['Total DM','Benchmark Exposure'] = sum_up(output, 'Benchmark Exposure')
    bm_dm = output.at['Total DM','Benchmark Exposure']
    bm_em = total - bm_dm
    output.at['Total EM','Benchmark Exposure'] = bm_em

    # Use float as data types and round to 1 decimal
    output = output.astype(float)
    output = output.round(1)
    
    # Update NA
    output.at['CAD', 'Strategic Hedges'] = pd.NA

    output.at['Total EM', 'Strategic Hedges'] = pd.NA
    output = output.drop("Hedgeable Exposure", axis=1)
    
    if (strategy == 'Other'):
        output = output.drop('Benchmark Exposure', axis=1)
        
    if (strategy == 'Active Asset Allocation'):
        output = output.drop('Benchmark Exposure', axis=1)
                
    return output.shape[1], output
```

```python
# Function for Allocation Monitor tab
def AM(strategy):
    # Filter the data based on the specified criteria
    df_filtered = df_op[(df_op["IPS_Strategy"] == strategy)]
        

    # Set up dataframe
    columns = ['RiskManager','RK']
    Clients = ['OPB','WSIBINS','WSIBLRI','WSIBPEN','PJPP','FSRA-PBGF']
    output = pd.DataFrame(columns=columns, index = Clients)

    for i in Clients:
        output.at[i, 'RiskManager'] = df_filtered.loc[df_filtered['Client_ID'] == i, 'Ownership_Multiplier_Updated'].sum()
      
    return output
```

```python
# Print out the table for this tab
def print_am(start_row, start_col):
            
    strategies = ["Global Infrastructure", "Global Credit", "Inflation Linked Bond",
            'Public Market Alternatives','Global Real Estate']
    with pd.ExcelWriter(new_file_name, engine = 'openpyxl', mode = 'a', if_sheet_exists = 'overlay') as writer:
        for i in range(len(strategies)):
            output = pd.DataFrame(AM(strategies[i]))
            
            output.to_excel(writer, sheet_name = 'Allocation Monitor', startrow = start_row, startcol = start_col, index = False, header = False)
            start_col += 2
               
```
